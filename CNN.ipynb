```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical
import numpy as np
```

```python
model=Sequential()
#ajout d'une couche de convolution
model.add(Conv2D(filters=3,kernel_size=(3,3),activation="relu",input_shape=(28,28,1)))
#ajout d'une couche de pooling
model.add(MaxPooling2D(pool_size=(2,2)))
#ajout d'une couche nouvelle couche de convolution + pooling
model.add(Conv2D(filters=3,kernel_size=(3,3),activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
#ajout d'une couche de flatten
model.add(Flatten())
#ajout des couches de MLP
model.add(Dense(units=12,activation="relu"))
model.add(Dense(units=20,activation="relu"))
model.add(Dense(units=8,activation="relu"))
#ajout de la couche de sortie
model.add(Dense(units=10,activation="softmax"))
#pour voir le model
model.summary()
```

```python
#chargement des données
(X_train,y_train),(X_test,y_test)=mnist.load_data()
print("X_train shape:",X_train.shape)
print("y_train shape:",y_train.shape)  
```

```python

#mise a l'echelle
X_train=X_train/255
X_test=X_test/255

y_train=to_categorical(y_train,len(np.unique(y_train)))
y_test=to_categorical(y_test, len(np.unique(y_test)))

print("X_train shape:",X_train.shape)
print("y_train shape:",y_train.shape)   
```

```python
#compilation du model
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])

#entrainement du model
model.fit(X_train,y_train,epochs=10,batch_size=20,validation_data=(X_test,y_test))
```

```python
acc_test=model.evaluate(X_test,y_test)[1]
print("accuracy sur le test:",acc_test)
```
